## 2. Modify model architecture and parameters

We've already init the default value for each parameter. They can be re-init by assigning new value for them. For example:

*parm.layers = [15]* - It means neural network has one 15-unit hidden layer.


```python
# Modify your model here and run this cell to update parameters
def your_model():
    parm.layers = [15] 
    parm.mnb_size = 150
    
    parm.n_epoch = 500
    parm.max_patience = 300
    
    parm.learning_rate = 0.01
    parm.momentum = 0.9
    
    parm.lamda = 0.00
    parm.run_dropout = False
    parm.run_batch_norm = False
    
    parm.act_func = parm.SIGMOID #parm.SIGMOID, parm.RELU, parm.TANH
    
your_model()
```

## 3. Train your model

*Train_Errs, Val_Errs, Model = run_training(plotshow = False)*

This function trains your model with your set-up parameters. 'plotshow' is boolean variable. Its default value is False. If you call *run_training()*, graph of errors thourgh n_epoch running times is not drawn.

Its output includes: Train_Errs, Val_Errs, Model

+ Train_Errs: include list of errors in training set thourgh n_epoch running times.

+ Val_Errs: include list of errors in validation set thourgh n_epoch running times.

+ *Model = stop, Ws, BN_parameter*

    - stop: stop epoch of Early Stopping.

    - Ws: a list of weight of model thourgh n_epoch running times.

    - BN_parameter: parameters for batch normalization. If don't use batch normalization, BN_parameter is None.

```python
Train_Errs_0, Val_Errs_0, Model_0 = run_training(True)
```

You can change value of parameters and re-train model.

```python
parm.act_func = parm.RELU
Train_Errs_1, Val_Errs_1, Model_1 = run_training(True)
```

## 4. Test your model

*Test_Errs = run_testing(Model)*

This function finds the error in testing set with model trained. Its input parameters: Model.

+ 'Model': the 3rd output result of run_training function. This variable stores training information.

Output of function is $R^2$ and MSE - Errors in testing.

```python
Test_Errs = run_testing(Model_1)
```
